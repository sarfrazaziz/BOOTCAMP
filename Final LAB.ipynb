```python
columns_to_drop = ['Consumer complaint narrative', 'Company public response','Consumer disputed?']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
plt.figure(figsize=(25,25))
sms.heatmap(data.isnull())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn
sns.set(style="whitegrid")

# Chart 1: Count of Complaints by Product
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Product', order=data['Product'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Product')
plt.ylabel('Count of Complaints')
plt.title('Count of Complaints by Product')
plt.show()

# Chart 2: Count of Complaints by State
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='State', order=data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Count of Complaints')
plt.title('Count of Complaints by State')
plt.show()

# Chart 3: Count of Complaints by Company Response
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='Company response to consumer', order=data['Company response to consumer'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Company Response to Consumer')
plt.ylabel('Count of Complaints')
plt.title('Count of Complaints by Company Response')
plt.show()

# Chart 4: Timely Response Rate
plt.figure(figsize=(6, 6))
sns.countplot(data=data, x='Timely response?', order=data['Timely response?'].value_counts().index)
plt.xlabel('Timely Response')
plt.ylabel('Count of Complaints')
plt.title('Timely Response Rate')
plt.show()

# Chart 5: Count of Complaints by Submitted via
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Submitted via', order=data['Submitted via'].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel('Submitted via')
plt.ylabel('Count of Complaints')
plt.title('Count of Complaints by Submission Method')
plt.show()
```
